```python
df1 = pd.read_csv('https://raw.githubusercontent.com/ShreyAgarwal11/NLP_Project/main/NQ_Dataset/nq_train1') # The URL points to a raw CSV file on GitHub
print(df1.shape)

df2 = pd.read_csv('https://raw.githubusercontent.com/ShreyAgarwal11/NLP_Project/main/NQ_Dataset/nq_train2') # The URL points to a raw CSV file on GitHub
print(df2.shape)

df3 = pd.read_csv('https://raw.githubusercontent.com/ShreyAgarwal11/NLP_Project/main/NQ_Dataset/nq_train3') # The URL points to a raw CSV file on GitHub
print(df3.shape)

df4 = pd.read_csv('https://raw.githubusercontent.com/ShreyAgarwal11/NLP_Project/main/NQ_Dataset/nq_train4') # The URL points to a raw CSV file on GitHub
print(df4.shape)

df5 = pd.read_csv('https://raw.githubusercontent.com/ShreyAgarwal11/NLP_Project/main/NQ_Dataset/nq_train5') # The URL points to a raw CSV file on GitHub
print(df5.shape)

df6 = pd.read_csv('https://raw.githubusercontent.com/ShreyAgarwal11/NLP_Project/main/NQ_Dataset/nq_train6') # The URL points to a raw CSV file on GitHub
print(df6.shape)
```
